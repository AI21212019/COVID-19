```python
import os
import pandas as pd
import us
from census import Census
import numpy as np
```

```python
#Selected HA census variables
df_variables= pd.read_csv(r'C:\Work_GovEx\COVID-19\New Data\census data download\Variable list.csv',encoding= 'unicode_escape')
df_variables.columns = ['var_name', 'source_variable', 'description']
#convert the table to a dictionary and create the full download list
variables = df_variables.set_index('var_name').to_dict()
#variables.keys()
rev_fields = {i[1]:i[0] for i in variables['source_variable'].items()}
varlist = list(rev_fields.keys())
#varlist
```

```python
ckey='e6abf23c528414abb0777c756b8df69a8f5b0481'
c = Census(ckey, year=2018)
Race_Eth_data = pd.DataFrame(c.acs5.state_county(varlist,Census.ALL,Census.ALL))
Race_Eth_data.head(2)
Race_Eth_data.rename(columns={'state':'state_fips', 'county':'county_fips'}, inplace=True)
```

```python
Race_Eth_data['FIPS']=Race_Eth_data['state_fips']+Race_Eth_data['county_fips']
Race_Eth_data['FIPS']=Race_Eth_data['FIPS'].apply(str).str.pad(width=5, side='left', fillchar='0')
Race_Eth_data.head(2)
```

```python
Race_Eth_data.rename(columns={'B02001_001E':'racePop_total', 'B02001_002E':'White alone', 'B02001_003E':'Black or African American alone', 
                              'B02001_004E':'American Indian and Alaska Native alone','B02001_005E':'Asian alone', 'B02001_006E':'Native Hawaiian and Other Pacific Islander alone', 
                              'B02001_007E':'Some other race alone', 'B02001_008E':'Two or more races','B03003_001E':'His_total','B03003_002E':'Not Hispanic or Latino origin', 'B03003_003E':'Hispanic or Latino Origin',},inplace=True)
```

```python
Race_Eth_data=Race_Eth_data[['NAME', 'racePop_total', 'White alone',
       'Black or African American alone',
       'American Indian and Alaska Native alone', 'Asian alone',
       'Native Hawaiian and Other Pacific Islander alone',
       'Some other race alone', 'Two or more races',
       'Not Hispanic or Latino origin', 'Hispanic or Latino Origin','FIPS']]
```

```python
Race_Eth_data.head(2)
```

```python
Age_data=pd.read_csv(r'C:\Work_GovEx\COVID-19\New Data\ACSST5Y2018.S0101_data_with_overlays_2020-05-07T004200.csv',low_memory=False)
Age_data1=Age_data[['GEO_ID', 'NAME','S0101_C01_002E','S0101_C01_020E', 'S0101_C01_005E','S0101_C01_006E',
                    'S0101_C01_007E','S0101_C01_008E','S0101_C01_009E','S0101_C01_010E','S0101_C01_011E', 
                     'S0101_C01_012E','S0101_C01_013E','S0101_C01_014E','S0101_C01_015E','S0101_C01_016E',
                    'S0101_C01_031E','S0101_C01_001E']]
Age_data1=Age_data1.iloc[1:]
Age_data1['FIPS']=Age_data1['GEO_ID'].str[-5:]
```

```python
Age_data1['Age_under15']=Age_data1['S0101_C01_002E'].astype(str).astype(int)+Age_data1['S0101_C01_020E'].astype(str).astype(int)
Age_data1['Age_15_24']=Age_data1['S0101_C01_005E'].astype(str).astype(int)+Age_data1['S0101_C01_006E'].astype(str).astype(int)
Age_data1['Age_25_34']=Age_data1['S0101_C01_007E'].astype(str).astype(int)+Age_data1['S0101_C01_008E'].astype(str).astype(int)
Age_data1['Age_35_64']=Age_data1['S0101_C01_009E'].astype(str).astype(int)+Age_data1['S0101_C01_010E'].astype(str).astype(int)+Age_data1['S0101_C01_011E'].astype(str).astype(int)+Age_data1['S0101_C01_012E'].astype(str).astype(int)+Age_data1['S0101_C01_013E'].astype(str).astype(int)+Age_data1['S0101_C01_014E'].astype(str).astype(int)
Age_data1['Age_65_74']=Age_data1['S0101_C01_015E'].astype(str).astype(int)+Age_data1['S0101_C01_016E'].astype(str).astype(int)
Age_data1['Age_over75']=Age_data1['S0101_C01_031E'].astype(str).astype(int)
Age_data1['Agetotal']=Age_data1['S0101_C01_001E'].astype(str).astype(int)
```

```python
Age_data1=Age_data1[[ 'FIPS', 'Age_under15', 'Age_15_24', 'Age_25_34',
       'Age_35_64', 'Age_65_74', 'Age_over75','Agetotal']]
```

```python
Age_data1.head()
```

```python
Race_Age_data=pd.merge(Race_Eth_data,Age_data1,how='left',on='FIPS')
```
